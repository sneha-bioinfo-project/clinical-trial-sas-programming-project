```python
import pandas as pd
from pathlib import Path

base = Path(
    "sdtm-adam-pilot-project/updated-pilot-submission-package/900172/m5/datasets/cdiscpilot01/tabulations/sdtm"
)

domains = ["dm", "ae", "lb", "vs", "ds"]

for d in domains:
    df = pd.read_sas(base / f"{d}.xpt")

    df.columns = [c.upper() for c in df.columns]

    out_file = f"Clinical-Trial-Pipeline/raw_data/{d.upper()}.csv"

    df.to_csv(out_file, index=False)

    print(f"{d.upper()}:", df.shape)
```

```python
import pandas as pd
import os

domains = ["DM","AE","LB","VS","DS"]

for domain in domains:

    df = pd.read_csv(
        f"Clinical-Trial-Pipeline/raw_data/{domain}.csv"
    )

    for col in df.columns:

        if df[col].dtype == object:

            df[col] = (
                df[col]
                .astype(str)
                .str.replace("b'", "", regex=False)
                .str.replace("'", "", regex=False)
            )

    df.to_csv(
        f"Clinical-Trial-Pipeline/raw_data/{domain}.csv",
        index=False
    )

    print(domain, "cleaned")
```
